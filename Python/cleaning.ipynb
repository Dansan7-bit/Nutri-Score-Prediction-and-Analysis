# Cleaning proccess

### Drop unnecessary columns

```python
df.drop(columns=['url', 'created_datetime', 'last_modified_datetime', 'last_updated_datetime', 'quantity', 'nova_group', 'additives_en', 'alcohol_100g', 'categories_en', 'categories_en.1', 'caffeine_100g', 'iron_100g'], inplace=True)
```

```python
df.shape
```

### Delete duplicates rows

```python
df = df.drop_duplicates()
df.shape
```

### Deleteing rows that contains NaN values in energy, fat, carbohydrates, protein and salt.

```python
nan_calories_rows = df[df[['energy-kcal_100g', 'fat_100g', 'carbohydrates_100g', 'proteins_100g', 'salt_100g']].isna().all(axis=1)]
nan_calories_rows
```

```python
df = df[~df[['energy-kcal_100g', 'fat_100g', 'carbohydrates_100g', 'proteins_100g', 'salt_100g']].isna().all(axis=1)]
df.shape
```

### Drop rows with energy above 1000 calories and below 0 (per 100 grams)

```python
df = df[(df['energy-kcal_100g'] > 0) & (df['energy-kcal_100g'] <= 1000)]
df.shape
```

### Dereived from this, applying a thresholds to the macronutrients columns.

```python
df = df[(df['fat_100g'] <= 111) & (df['proteins_100g'] <= 250) & (df['carbohydrates_100g'] <= 250)]
df.shape
```

### Calorie intake can't be less than it's macronutrients. therefore, a column with the expected calore intake will be created based on the macronutrients and then filtered by it so the calories column will must be equal or greater than the expected calorie intake.

```python
df['expected_min_calories'] = (df['proteins_100g'] * 4) + (df['carbohydrates_100g'] * 3.5) + (df['fat_100g'] * 9)
```

```python
df = df[df['energy-kcal_100g'] >= df['expected_min_calories']]
df.shape
```

### Remove Rows Without product name or product with 'unknown' name

```python
df['product_name'] = df['product_name'].replace('unknown', np.nan)
df = df.dropna(subset=['product_name'])
df.shape
```

### Delete rows with NaN or unknown name values in pnns groups

```python
columns_to_check = ['pnns_groups_1', 'pnns_groups_2']
df[columns_to_check] = df[columns_to_check].replace('unknown', np.nan)

# Drop rows with NaN values in the specified columns
df.dropna(how='all', subset=columns_to_check, inplace=True)
df.shape
```

```python
df.isnull().sum()
```

### Delete rows where NaN values are in the specified columns.

```python
df.dropna(subset=['nutriscore_grade', 'countries_en', 'ecoscore_grade'], inplace=True)
```

### Delete rows where their nutri-score is 'not-applicable'.

```python
df = df[df['nutriscore_grade'] != 'not-applicable']
```

### Rename columns.

```python
df.rename(columns={
    'nutriscore_grade': 'nutriscore',
    'product_name': 'product',
    'countries_en': 'country',
    'pnns_groups_1': 'primary_category',
    'pnns_groups_2': 'secondary_category',
    'ecoscore_grade': 'ecoscore',
    'energy-kcal_100g': 'energy_kcal_100g',
    'fat_100g': 'fat_100g',
    'saturated-fat_100g': 'saturated_fat_100g',
    'unsaturated-fat_100g': 'unsaturated_fat_100g',
    'monounsaturated-fat_100g': 'mono_unsaturated_fat_100g',
    'polyunsaturated-fat_100g': 'poly_unsaturated_fat_100g',
    'trans-fat_100g': 'trans_fat_100g',
    'cholesterol_100g': 'cholesterol_100g',
    'carbohydrates_100g': 'carbs_100g',
    'sugars_100g': 'sugars_100g',
    'fiber_100g': 'fiber_100g',
    'proteins_100g': 'protein_100g',
    'salt_100g': 'salt_100g',
    'sodium_100g': 'sodium_100g',
    'fruits-vegetables-nuts-estimate-from-ingredients_100g': 'fruits_veg_nuts_estimate_100g',
    'expected_min_calories': 'expected_min_calories'
}, inplace=True)
```
